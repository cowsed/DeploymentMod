```python
xs = [1,2,3,4]
ys = [2,4,6,8]
```

```python
N = len(xs)
```

```python
Exy = sum(x * y for (x,y) in zip(xs, ys))
Ex2 = sum(x * x for x  in xs)

Ex = sum(xs)
Ey = sum(ys)
```

```python
m = (N * Exy - Ex * Ey) / (N * Ex2 - Ex*Ex)
```

```python
b = (Ey - m * Ex)/N
```

```python
print(f"{m}x + {b}")
```
